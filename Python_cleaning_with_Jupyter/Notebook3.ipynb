```python
import pandas as pd

df = pd.read_csv('customer_data.csv')

df.head()
```

```python
def print_invalid_count(df):
    print("invalid email")
    invalid_email_count = df[df["email"] == "invalid_mail"].shape[0]
    invalid_age_count = df[df["age"] == "invalid_age"].shape[0]

    print(f"invalid emails: {invalid_email_count}")
    print(f"invalid ages: {invalid_age_count}")

print_invalid_count(df)
```
